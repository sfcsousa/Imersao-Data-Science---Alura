### Desafio 01: Investigar o por que a classe tratamento é tão desbalanceada

É visto que há um desbalanceamento na série tratamento pelo falo que de as amostras com_controle são amostras de controle, que servem com o parâmetro de comparação com a reação das outras amostras.

```python
print(f"Total de dados {len(base_dados['id'])}\n")
print(f"Quantidade de drogas {len(base_dados.groupby(['droga', 'tratamento']).count()['id'])}\n")
display(base_dados.query('tratamento == "com_controle"').value_counts('droga'))
print()
print('Visto, então, que a droga cacb2b860 foi a utilizada como parâmetro base de controle do experimento')
print()
display(base_dados.query('droga == "cacb2b860"').value_counts('tratamento'))
print()
display(base_dados.query('droga != "cacb2b860"').value_counts('tratamento'))
print()
display(base_dados.query('tratamento != "com_controle"').value_counts('droga'))
```

### Desafio 05 Refletir sobre a manipulação dos tamanhos das visualizações.

A importância na visualização de um gráfico está completamente relacionada a comunicação entre o loucutor e ouvinte. O gráfico ele deve transmitir a informação tratada e analisada de forma que o ouvinte, a audiência capte a informação apresentada, como em uma conversa que flui normalmente. 

Atributos como cor, tamanho da fonte, estilo da fonte, legendas, espaçamento, contribuem para facilitar(ou não) essa comunicação e transmissão da informação.
